```python
%matplotlib inline #notebook을 실행한 브라우저에서 바로 그림을 볼 수 있게 해주는 것

import pandas as pd #import 모듈명 as 별명
import networkx as nx 
import matplotlib.pyplot as plt
```

```python
import matplotlib.font_manager as fm

from matplotlib import rc #from 모듈 import 함수

font_name = fm.FontProperties(fname="c:/Windows/Fonts/malgun.ttf").get_name()

rc('font', family=font_name)
```

```python
df = pd.read_csv('C:/Users/nayeonkim/Desktop/featuring_bts.csv') #2020년 bts가 피처링한 가수의 피처링한 가수 파도타기 데이터 (직접만듦)
df.head() 
```

```python
g = nx.from_pandas_edgelist(df, source='from', target='to',create_using = nx.DiGraph()) 
plt.figure(3,figsize=(12,12)) 
nx.draw(g, with_labels = True, font_family=font_name,font_size=15)
```
